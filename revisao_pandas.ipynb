```python
import pandas as pd

# df = dataframe = tabela
vendas_df = pd.read_csv("Vendas_Chocolate.csv", sep=";")
display(vendas_df)
```

```python


# valores vazios
# 2 tipos: linhas com valores vazios e colunas com muitos valores vazios
vendas_df = vendas_df.drop(columns="Data de Hoje")
vendas_df = vendas_df.dropna() # exclui linhas que têm algum valor vazio

# valores em formatação errada
# object -> textos
# float -> numeros com casa decimal
# datetime -> datas
vendas_df["Data"] = pd.to_datetime(vendas_df["Data"], dayfirst=True)
# Valor -> $5,320.00
vendas_df["Valor"] = vendas_df["Valor"].str.replace("$", "").str.replace(",", "")
vendas_df["Valor"] = vendas_df["Valor"].astype(float)


# valores duplicados
vendas_df = vendas_df.drop_duplicates()

# reset no indice da base de dados
vendas_df = vendas_df.reset_index(drop=True)

print(vendas_df.info()) # identificar problemas na base de dados
display(vendas_df)
```

```python
display(vendas_df.describe()) # cria uma descrição estatística das colunas numéricas

```

```python
# feature engineering
vendas_df["Mes"] = vendas_df["Data"].dt.to_period("M")
# dt = datetime
vendas_df["Ano"] = vendas_df["Data"].dt.year
display(vendas_df)

```

```python
# analise de crescimento de vendas
import plotly.express as px

# temporal (dias/mes/ano)
vendas_mes = vendas_df[["Mes", "Valor"]]
vendas_mes = vendas_mes.groupby("Mes", as_index=False).sum()
display(vendas_mes)

# grafico
grafico = px.bar(vendas_mes, x=vendas_mes["Mes"].astype(str), y="Valor")
grafico.show()
```

```python
# analise de volume e ticket medio anual
vendas_ano = vendas_df[["Ano", "Valor", "Caixas Enviadas"]]
vendas_ano = vendas_ano.groupby("Ano", as_index=False).sum()
vendas_ano["Ticket Medio"] = vendas_ano["Valor"] / vendas_ano["Caixas Enviadas"]

vendas_ano["% Crescimento Valor"] = vendas_ano["Valor"].pct_change()

display(vendas_ano.style.format({
    "Valor": "${:,.2f}",
    "Caixas Enviadas": "{:,.0f}",
    "Ticket Medio": "${:,.2f}",
    "% Crescimento Valor": "{:,.1%}",
}))
```
